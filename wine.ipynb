```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
wine = datasets.load_wine()
X = wine.data
y = wine.target
```

```python
pca = PCA(n_components=2)
X_red = pca.fit_transform(X)
```

```python
plt.scatter(X_red[:, 0], X_red[:, 1], c=y, edgecolor="none", alpha=0.5, cmap=plt.cm.get_cmap('nipy_spectral', 3))
plt.colorbar(label="wine", ticks=range(3))
```
