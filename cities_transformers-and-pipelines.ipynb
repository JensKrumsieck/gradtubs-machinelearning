```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
df = pd.DataFrame({'city': ['London', 'London', 'Paris', 'Sallisaw', np.nan],
                   'rating': [5, 3, 4, 5, 6]})
df
```

### Manual Way


```python
si = SimpleImputer(strategy="most_frequent")
cities_imputed = si.fit_transform(df[["city"]].values)
```

```python
ohe = OneHotEncoder(handle_unknown="ignore")
ohe.fit_transform(cities_imputed).toarray()
```

### Pipelines


```python
from sklearn.pipeline import Pipeline
si_ohe = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy="most_frequent")),
        ('onehot', OneHotEncoder())
    ]
)
si_ohe.fit_transform(df[["city"]].values).toarray()
```

### Transformers ðŸ¤–

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
```

```python
ct = ColumnTransformer(transformers=[
    ("preprocess_city", Pipeline(
        steps=[
            ('imputer', SimpleImputer(strategy="most_frequent")),
            ('onehot', OneHotEncoder())
        ]
    ), ["city"]),
    ("preprocess_rating", MinMaxScaler(), ["rating"])
])
```

```python
ct.fit_transform(df)
```
